# Install TypeScript kernell for Jupyter

![TypeScript](TypeScript.png) On va faire une revision du language [TypeScript](https://fr.wikipedia.org/wiki/TypeScript) avec [les bases de TypeScript](https://www.typescriptlang.org/docs/handbook/2/basic-types.html).  

Pour tester les exemples lors de la lecture j'ai choisi d'utiliser l'environnement Jupyter Notebook avec VSCode. Je peut aussi me servir du terminal si besoin mais je préfère le faire sur une fenêtre séparée.  

On va installer l'environnement. Je vous ferais pas l'affront de vous montrer comment installer VSCodium, l'install de l'environnement nécessaire pour executer des notebooks se trouve [ici](https://github.com/rnek0/myJupyterVenv).  

Dans un premier temps il nous faut installer TypeScript, pour cela il nous faut **npm** (le gestionnaire de paquets pour JavaScript), il est intéressant de savoir que npm a été acheté par Microsoft et que TypeScript est également crée par la même entreprise, par le créateur du language C# [Anders Hejlsberg](https://fr.wikipedia.org/wiki/Anders_Hejlsberg), alors que [JavaScript](https://fr.wikipedia.org/wiki/JavaScript) existe depuis bien avant et fut crée par [Brendan Eich](https://fr.wikipedia.org/wiki/Brendan_Eich) lorsqu'il qu'il travaillait chez Netscape. 

```bash
sudo npm install -g typescript
```

Il se peut que vous ayez a mettre a jour npm moi a ce jour je le fais avec cette version: 

```bash
sudo npm install -g npm@9.6.3
```

N'hésitez pas a bien relire les commentaires lors des installs, moi par exemple j'ai un package **deprecated** et on me conseille de faire une mise à jour

Après le socle, on va installer l'environnement Jupiter Notebook et le kernell pour TypeScript.

## Lancer les notebooks

Une fois l'environnement installé, on peut au choix travailler avec VSCode ou bien regarder sur un navigateur en même temps. Il y a même la possibilité de faire des retouches sur l'interface web.  
Pour ce faire on lance les notebooks comme ceci :

```bash
jupyter notebook
```

Cela va démarrer le serveur et on pourra aller consulter les notebooks sur le browser.


## Configuration de TEX pour la generation de pdf's

Pour installer les outils pour la generation de pdf con TEX

```bash
wget https://mirror.ctan.org/systems/texlive/tlnet/install-tl-unx.tar.gz # o curl en lugar de wget
zcat < install-tl-unx.tar.gz | tar xf -
cd install-tl-*
perl ./install-tl --no-interaction # como root o con destino grabable
```

Puis ne pas oublier de sourcer les PATH :

```bash
# TEX INSTALL
export PATH=/usr/local/texlive/2023/bin/x86_64-linux:$PATH
export MANPATH=/usr/local/texlive/2023/texmf-dist/doc/man:$MANPATH
export INFOPATH=/usr/local/texlive/2023/texmf-dist/doc/info:$INFOPATH
```
